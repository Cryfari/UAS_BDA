```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import association_rules
from mlxtend.frequent_patterns import fpgrowth
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/Cryfari/TugasBigDataAnalitik/main/anime-dataset-2023.csv", usecols=['Genres'])
df.head()
```

```python
df.shape
```

```python
df = df[df['Genres'] != 'UNKNOWN']
df.shape
```

```python
data = list(df["Genres"].apply(lambda x:x.split(',')))
data
```

```python
for i in data:
  for j in range(len(i)):
    i[j] = i[j].strip()
data
```

```python
te = TransactionEncoder()
te_array = te.fit(data).transform(data)
df = pd.DataFrame(te_array, columns=te.columns_)
df
```

```python
frequent_itemsets_fp=fpgrowth(df, min_support=0.02, use_colnames=True)
```

```python
frequent_itemsets_fp.shape
```

```python
frequent_itemsets_fp.sort_values(by='support', ascending=False)[0:10]
```

```python
rules_fp = association_rules(frequent_itemsets_fp, metric="confidence", min_threshold=0.5)
```

```python
rules_fp.sort_values(by='confidence', ascending=False)
```
